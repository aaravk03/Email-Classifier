## Evaluating Classifiers

In our models, we evaluate the accuracy of the training set, which may provide a misleading accuracy measure. In Project B1, we calculated various metrics to consider other ways of evaluating a classifier, in addition to overall accuracy. Below is a reference to those concepts.

Presumably, our classifier will be used for **filtering**, or preventing messages labeled `spam` from reaching someone's inbox. There are two kinds of errors we can make:
- **False positive (FP)**: A ham email gets flagged as spam and filtered out of the inbox.
- **False negative (FN)**: A spam email gets mislabeled as ham and ends up in the inbox.

To be clear, we label spam emails as 1 and ham emails as 0. These definitions depend both on the true labels and the predicted labels. False positives and false negatives may be of differing importance, leading us to consider more ways of evaluating a classifier in addition to overall accuracy:

**Precision**: Measures the proportion of emails flagged as spam that are actually spam. Mathematically, $\frac{\text{TP}}{\text{TP} + \text{FP}}$.

**Recall**: Measures the proportion  of spam emails that were correctly flagged as spam. Mathematically, $\frac{\text{TP}}{\text{TP} + \text{FN}}$.

**False positive rate**: Measures the proportion  of ham emails that were incorrectly flagged as spam. Mathematically, $\frac{\text{FP}}{\text{FP} + \text{TN}}$.

3. We fit each model to some training data `X_train` and `Y_train` using `grid_result = grid.fit(X_train, Y_train)`.
4. Indexing into `grid_result.cv_results_` with a particular metric (in this case, `mean_test_score`), we get an array with the scores corresponding to each of the models. `grid_result.cv_results_['mean_test_score']`.

# Question 5: ROC Curve

In most cases, we won't be able to get 0 false positives and 0 false negatives, so we have to compromise. For example, in the case of cancer screenings, false negatives are comparatively worse than false positives — a false negative means that a patient might not discover that they have cancer until it's too late. In contrast, a patient can receive another screening for a false positive.

Recall that logistic regression calculates the probability that an example belongs to a particular class. To classify an example, we say that an email is spam if our classifier gives it $\ge 0.5$ probability of being spam. However, **we can adjust that cutoff threshold**. We can say that an email is spam only if our classifier gives it $\ge 0.7$ probability of being spam, for example. This is how we can trade off false positives and false negatives.

The Receiver Operating Characteristic (ROC) curve shows this trade-off for each possible cutoff probability. In the cell below, plot an ROC curve for your final classifier (the one you use to make predictions for Gradescope) on the training data. [Lecture 23](https://ds100.org/sp24/lecture/lec23/) may be helpful.

**Hint**: You'll want to use the `.predict_proba` method [(documentation)](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html#sklearn.linear_model.LogisticRegression.predict_proba) for your classifier instead of `.predict` to get probabilities instead of binary predictions.

```python
spam_probabilities = model.predict_proba(X_train)[:, 1]
fpr, tpr, thresholds = roc_curve(Y_train, spam_probabilities)
plt.figure(figsize=(10, 8))
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Email Classification')
plt.legend(loc="lower right")
plt.show()
```

```python
# Simple model introduced at the start of this notebook. Just pay attention to the features.
some_words = ['drug', 'bank', 'prescription', 'memo', 'private']

X_train = words_in_texts(some_words, train['email'])
Y_train = np.array(train['spam'])

simple_model = LogisticRegression()
simple_model.fit(X_train, Y_train);
```

Pick an email from the training set and assign its index to `email_idx`. Then, find **one** feature used in `simple_model` such that **removing** it changes how that email is classified. Assign this feature to  `feature_to_remove`.

```python
feature_to_remove = 'bank'

mask_contains_bank = words_in_texts([feature_to_remove], train['email'])[:, 0]

emails_with_bank = train.loc[mask_contains_bank]

X_with_bank = words_in_texts(some_words, emails_with_bank['email'])
predictions_with_bank = simple_model.predict(X_with_bank)

emails_with_bank = emails_with_bank.assign(prediction=predictions_with_bank)

spam_emails_with_bank = emails_with_bank[(emails_with_bank['spam'] == 1)]

ham_emails_with_bank = emails_with_bank[(emails_with_bank['spam'] == 0)]

print(f"Number of emails containing 'bank' labeled as spam: {len(spam_emails_with_bank)}")
print(f"Number of emails containing 'bank' labeled as ham: {len(ham_emails_with_bank)}")
```

```python
train['email'] = train['email'].astype(str)
mask_contains_bank = train['email'].str.contains(r'\b{}\b'.format(feature_to_remove), case=False, regex=True)

emails_with_bank_and_spam = train[mask_contains_bank & (train['spam'] == 1)]

if not emails_with_bank_and_spam.empty:
    X_emails_with_bank_and_spam = words_in_texts(some_words, emails_with_bank_and_spam['email'])
    predictions = simple_model.predict(X_emails_with_bank_and_spam)

    mismatches = predictions != emails_with_bank_and_spam['spam']
    mismatched_emails = emails_with_bank_and_spam[mismatches]

    print(f"Number of spam emails containing 'bank': {len(emails_with_bank_and_spam)}")
    print(f"Number of spam emails where prediction mismatches the label: {len(mismatched_emails)}")
else:
    print("No spam emails containing 'bank' were found.")
```

```python
reduced_features = [word for word in some_words if word != feature_to_remove]

X_train_reduced = words_in_texts(reduced_features, train['email'])

reduced_model = LogisticRegression()
reduced_model.fit(X_train_reduced, train['spam'])

X_spam_with_bank_reduced = words_in_texts(reduced_features, emails_with_bank_and_spam['email'])
new_predictions = reduced_model.predict(X_spam_with_bank_reduced)

changed_classifications_mask = new_predictions != emails_with_bank_and_spam['spam']
emails_with_changed_classification = emails_with_bank_and_spam[changed_classifications_mask]

print(f"Number of spam emails containing 'bank' whose classification changed after removing 'bank': {len(emails_with_changed_classification)}")

if not emails_with_changed_classification.empty:
    print("Indices of emails with changed classification:", emails_with_changed_classification.index.tolist())
```

```python
email_idx = 68

prob_spam = simple_model.predict_proba(X_train)[:, 1]
initial_prob = prob_spam[email_idx]
initial_class = "spam" if np.round(initial_prob) else "ham"
print(f"\nPredicted probability of being spam: {np.round(initial_prob*100, 2)}%")
print("\nEmail:\n" + train.loc[email_idx]["email"])
```

```python
feature_to_remove = 'bank'

changed_words = some_words.copy()
changed_words.remove(feature_to_remove)

changed_model = LogisticRegression()
X_changed = words_in_texts(changed_words, train['email'])
y = train['spam']
changed_model.fit(X_changed, y)
changed_prob = changed_model.predict_proba(X_changed[[email_idx]])[:,1][0]
changed_class = "spam" if np.round(changed_prob) else "ham"

print(f"Initially classified as {initial_class} (Probability: {np.round(initial_prob*100, 2)}%)")
print(f"Now classified as {changed_class} (Probability: {np.round(changed_prob*100, 2)}%)")
```

### Question 7e

As a data scientist, why might having an interpretable model be useful when moderating content online?

A data scientist involved in online content moderation will find it very helpful to have an interpretable model for the following reasons:

- Accountability: Data scientists and the platforms they support can comprehend and provide an explanation for the decisions made when using interpretable models. Being able to provide an explanation for a post's flagging or removal in the context of content moderation is essential for upholding user confidence and adhering to legal requirements. When users are aware of the justification for material removal, they are more likely to be treated properly.
- Transparency: To guarantee that the guidelines are followed properly and consistently, transparency is crucial in the content moderation process. The clear enforcement of moderation policies can be elucidated through the use of an interpretable model. This openness is crucial for external oversight by civil society organizations and regulatory bodies, in addition to users.
- Refinement of Moderation Policy: Interpretable models aid in determining which content characteristics influence a given moderation decision. This knowledge is crucial for enhancing the precision of content moderation and detection systems. Data scientists can reduce errors, such false positives and false negatives, by adjusting thresholds or features by analyzing the judgments made by the models.
- Handling Appeals: An interpretable model makes the review process more transparent when users disagree with moderation choices. It helps to analyze the decision and make sure it was accurate in accordance with the platform's standards and policies if the logic behind the model is easily accessible and understandable.
- Informative Comments for Users: Using an interpretable model, platforms can inform users in detail about the parts of their material that did not comply with guidelines. Users may benefit from this in the long run by learning about and being able to adhere to community standards.

In conclusion, the capacity of a model to be interpreted in the context of online content moderation facilitates improved governance, improves user experience, and guarantees the equitable implementation of community norms. The integrity of the platform and user engagement depend on these elements.
